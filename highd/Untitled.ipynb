```python
import pandas as pd
import numpy as np
```

```python
df2 = pd.DataFrame()
name_dataset = 'highd-dataset-v1.0'
for i in range(1, 61):
    num_csv = f"0{i}" if len(str(i)) != 2 else str(i)
    
    track_csv = f"../../lololol/data/datasets/{name_dataset}/data/" + num_csv + "_tracks.csv"
    labels_csv = f'output/{name_dataset}_{num_csv}_dataset_label.csv'
    track_dist = f'output2/{name_dataset}_{num_csv}_dataset_concat_dist.csv'
    track = pd.read_csv(track_csv)
    labels = pd.read_csv(labels_csv)
    dist = pd.read_csv(track_dist)
    df = labels.reset_index().merge(dist.reset_index(), on=['id', 'frame'])
    df = df.reset_index().merge(track.reset_index(), on=['id', 'frame'])
    df2 = pd.concat([df2, df], ignore_index=True)
```

```python
column = ['precedingXAcceleration', 'followingXAcceleration', 'followingXVelocity',
                  'leftPrecedingXAcceleration', 'leftPrecedingXVelocity',
                  'leftAlongsideXAcceleration', 'leftAlongsideXVelocity',
                  'leftFollowingXAcceleration', 'leftFollowingXVelocity',
                  'rightPrecedingXAcceleration', 'rightPrecedingXVelocity',
                  'rightAlsongsideXAcceleration', 'rightAlsongsideXVelocity',
                  'rightFollowingXAcceleration', 'rightFollowingXVelocity',
                  'precedingDist', 'followingDist', 'leftPrecedingDist',
                  'leftAlongsideDist', 'leftFollowingDist',
                  'rightPrecedingDist', 'rightAlsongsideDist',
                  'rightFollowingDist', 'x', 'y', 'xVelocity', 'xAcceleration', 'yAcceleration',
          'frontSightDistance', 'backSightDistance', 'dhw', 'thw', 'ttc', 'precedingXVelocity', 'label', 'class'
                  ]
```

```python
df2 = df2[column]
```

```python
df3 = df2[df2['class'] == 'Car']
```

```python
Y = df3['label']
X = df3.drop(['label', 'class'], axis=1)
```

```python
params = {'criterion': ['gini', 'entropy'],
         'max_depth': range(1,10), 
         'min_samples_split': range(1,10),
         'min_samples_leaf': range(1,5)}
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()


```

```python
clf = GridSearchCV(decision_tree, param_grid=params, cv=4, verbose=1, n_jobs=2)
```

```python
clf.fit(X, Y)
```

```python
import json
with open(f'car-param.json', 'w') as f:
    json.dump(clf.best_params_, f)
```

```python
df3 = df2[df2['class'] == 'Truck']
Y = df3['label']
X = df3.drop(['label', 'class'], axis=1)
decision_tree = DecisionTreeClassifier()
clf = GridSearchCV(decision_tree, param_grid=params, cv=4, verbose=1, n_jobs=-1)
```

```python
clf.fit(X, Y)
```

```python
import json
with open(f'truck-param.json', 'w') as f:
    json.dump(clf.best_params_, f)
```
